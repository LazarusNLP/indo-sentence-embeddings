```python
import pandas as pd

# some rows contain >7 columns, so we need to ignore those
lines = open("../data/sts-test.csv").read().split("\n")
columns = ['domain','data','type','score','correlation','text_1','text_2']
df = pd.DataFrame([dict(zip(columns, line.split("\t")[:7])) for line in lines[:-1]])
df
```

```python
from tqdm.contrib.concurrent import thread_map
import mtranslate

translate = lambda text: mtranslate.translate(text, "id", "en")
df["text_1"] = thread_map(translate, df["text_1"])
df["text_2"] = thread_map(translate, df["text_2"])
```

```python
df
```

```python
df.to_csv("../data/sts-test-ind.csv", index=False)
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df, split="test")
dataset.push_to_hub("LazarusNLP/stsb_mt_id", private=True)
```
